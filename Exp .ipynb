```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(path + "/books.csv")
df.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ax = plt.axes()

sns.heatmap(df.isnull().transpose(), cbar=False, cmap='viridis', ax=ax)
plt.title("Missing Data Heatmap")
plt.show()
```

```python
import numpy as np

df["missing_descriptions"] = np.where(df["description"].isna(),1,0)
df["age_of_books"] = 2024 - df["published_year"]
```

```python
df.head(10)
```

```python
column_of_interests = ["num_pages","average_rating", "missing_descriptions","age_of_books"]

correlation_matrix = df[column_of_interests].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(10,10))

sns.heatmap(correlation_matrix, cmap="coolwarm", annot=True)
plt.show()
```

```python
df[
    (df["num_pages"].isna()) |
    (df["published_year"].isna()) |
    (df["average_rating"].isna()) |
    (df["description"].isna())
]
```

```python
df.shape
```

```python
new_df = df[
    ~(df["num_pages"].isna()) &
    ~(df["published_year"].isna()) &
    ~(df["average_rating"].isna()) &
    ~(df["description"].isna())
]
```

```python
new_df.shape
```

```python
new_df.head(10)
```

```python
new_df["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
new_df["words_in_description"] = new_df["description"].str.split().str.len()
new_df
```

```python
new_df.loc[new_df["words_in_description"].between(5,14), "description"]
```

```python
new_df
```

```python
book_missing_25_words = new_df[new_df["words_in_description"] >= 25]
book_missing_25_words
```

```python
book_missing_25_words["titles_and_subjects"] = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],
    book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)
)
book_missing_25_words
```

```python
book_missing_25_words["tagged_descriptions"] = book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join, axis=1)
book_missing_25_words
```

```python
book_missing_25_words.drop(
    ["description","words_in_description","missing_descriptions","age_of_books"],axis=1
    ).to_csv("Books.csv",index=False)
```
